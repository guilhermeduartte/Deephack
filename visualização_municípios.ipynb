```python
import matplotlib.pyplot as plt
import mplleaflet
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
munis = pd.read_csv('datasets/municipios.csv')
```

```python
munis['Longitude']
```

```python
plt.figure(figsize=(8,15))

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

sizes = scaler.fit_transform(munis['Estimativa 2019'].values.reshape(-1,1))
sizes = sizes.reshape(len(sizes))

munis.plot.scatter('Longitude', 'Latitude', s = sizes*200, c = 'coral', alpha = 0.7)
mplleaflet.display()
```
